```python
import datetime
import time
import pyupbit
import ccxt
from google.oauth2 import service_account
import pandas_gbq

binance = ccxt.binance()
board = ['GXS', 'LUNA', 'FIL', 'SUN', 'USDP', 'RVN', 'OXT', 'INJ', 'NEAR', 'MASK', 'SNX', 'FOR', 'RSR']

def get_data(b):
    if b == 'GXS':
        ticker = binance.fetch_ticker('GXS/USDT')
        data = pyupbit.get_orderbook(tickers="BTC-GXC")

        upbit = data[0]['orderbook_units'][0]
        
        return [ticker, upbit]
    
    # 업비트랑 바이낸스 이름이 같은 것
    else :
        ticker = binance.fetch_ticker(b+'/USDT')
        data = pyupbit.get_orderbook(tickers="BTC-"+b)

        upbit = data[0]['orderbook_units'][0]
        
        return ticker, upbit
        
# date = '2021-01-01'

with open('C:/Users/choi/Desktop/data/21-12-16.csv', 'a') as f:
# with open('./'+date+'.csv', 'a') as f:
    f.writelines(['datetime,', 'timestamp,'])
    
    for b in board:
        f.writelines([b+','])
        
    f.write('Category\n')
    

while True:
    with open('C:/Users/choi/Desktop/data/21-12-16.csv', 'a') as f:
#     with open('./'+date+'.csv', 'a') as f:
        
        try:
            now = datetime.datetime.now()
            nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
#             date = now.strftime('%Y-%m-%d')
            nowTimeStamp = now.timestamp()
            
#             if nowTimeStamp % 86400 < 5  : # 24시간
                
#             else :
#                 pass
            
            to_write = []
#             to_write = [nowDatetime, str(nowTimeStamp)]
            
#             f.writelines([nowDatetime, ',', str(nowTimeStamp), ','])

            for b in board:
                
                # 불러오는데 0.1~0.3초 걸림
                ticker, upbit = get_data(b)
                
                to_write.extend([str(ticker['close']), str(upbit['ask_price']), str(upbit['bid_price'])])

            for cate in range(3):
                f.writelines([nowDatetime, ',', str(nowTimeStamp), ','])
                
                for idx in range(len(board)):
                    
                    f. writelines([to_write[3*idx + cate], ','])
                
                if cate == 0:
                    f.write('binance\n')
                elif cate == 1:
                    f.write('up_ask\n')
                else :
                    f.write('up_bid\n')
            
        except TypeError as e :
            now = datetime.datetime.now()
            nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
            print(nowDatetime, e)
```
