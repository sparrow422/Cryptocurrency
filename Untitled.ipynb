```python
default = 0
flag = False

while True:
    ticker = binance.fetch_ticker('GXS/USDT')
    if default != ticker['close'] :
        if flag :
            # 1% 올랐을 때
            if ticker['close'] > default * 1.01 :
                # buy upbit
                
            default = ticker['close']
                
        else:
            default = ticker['close']
            flag = True
            
    print(ticker['open'], ticker['high'], ticker['low'], ticker['close'])
```

```python
import datetime
import time
import pyupbit
import ccxt

binance = ccxt.binance()

with open('C:/Users/choi/Desktop/data/21-12-13.csv', 'a') as f:
    f.writelines(['datetime,','timestamp,', 'GXS_binance,', 'GXS_up_ask,', 'GXS_up_bid', 
                  'LUNA_binance,', 'LUNA_up_ask,', 'LUNA_up_bid''\n'])
    
with open('C:/Users/choi/Desktop/data/21-12-13.csv', 'a') as f:
    while True:
        try:
            now = datetime.datetime.now()
            nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
            nowTimeStamp = now.timestamp()
            
            ticker = binance.fetch_ticker('GXS/USDT')
            data = pyupbit.get_orderbook(tickers="BTC-GXC")

            upbit = data[0]['orderbook_units'][0]

        #     print(nowDatetime, nowTimeStamp, ticker['close'], upbit['ask_price'], upbit['bid_price'])

            f.writelines(','.join([nowDatetime, str(nowTimeStamp),str(ticker['close']), str(upbit['ask_price']), str(upbit['bid_price'])]))
            f.write(',')
            
            ticker = binance.fetch_ticker('LUNA/USDT')
            data = pyupbit.get_orderbook(tickers="BTC-LUNA")

            upbit = data[0]['orderbook_units'][0]

            f.writelines(','.join([str(ticker['close']), str(upbit['ask_price']), str(upbit['bid_price'])]))
            f.write('\n')
            
        except TypeError as e :
            now = datetime.datetime.now()
            nowDatetime = now.strftime('%Y-%m-%d %H:%M:%S')
            print(nowDatetime, e)
```
